---
Una juguetería tiene mucho éxito en dos de sus productos: payasos y muñecas. Suele hacer venta por correo y
la empresa de logística les cobra por peso de cada paquete así que deben calcular el peso de los payasos y muñecas
que saldrán en cada paquete a demanda. Cada payaso pesa 112 g y cada muñeca 75 g. Escribir un programa que lea el número
de payasos y muñecas vendidos en el último pedido y calcule el peso total del paquete que será enviado


```python
# Definicion de una funcion con 2 parametros
def ventas(p, m):
  peso_payaso = 112
  peso_munieca = 75

  # calculando la cantidad de payasos y su peso y d muñecas
  calcular_payaso = peso_payaso * p
  calcular_munieca = peso_munieca * m

  print(f'El peso total de los payasos es: {calcular_payaso} g.')
  print(f'El peso total de las muñecas es: {calcular_munieca} g.')
  #calculando el peso total de ambos juguetes
  print(f'\nel peso total del paquete a enviar de los juguetes es de {calcular_payaso * calcular_munieca} g.')


# variables de ingreso de valores por consola
numero_payaso = int(input('Ingrese total de payasos: '))
numero_munieca = int(input('Ingrese total de nuñecas: '))

# instancia de ventas con 2 argumentos
ventas(numero_payaso, numero_munieca)
```

Imagina que acabas de abrir una nueva cuenta de ahorros que te ofrece el 4% de interés al año. Estos ahorros debido
a intereses, que no se cobran hasta finales de año, se te añaden al balance final de tu cuenta de ahorros. Escribir un
programa que comience leyendo la cantidad de dinero depositada en la cuenta de ahorros, introducida por el usuario. Después
el programa debe calcular y mostrar por pantalla la cantidad de ahorros tras el primer, segundo y tercer años. Redondear cada
cantidad a dos decimales

```python
# fincion de validacion numerico
def validacion(n):
  # bucl infinito
  while True:
    try:
      # ingresa la cantidad de n
      cant = int(input(n))
      return cant
    except:
      print('Error! Ingresar solo valores númericos.')


def cuenta_ahorros():
  # llamado de la funcio VALIDACION con dicho argumento
  usuario = validacion('Ingrese su deposito. ')

  # comprueba calculo
  calculo = usuario * 4 / 100
  primer_anio = calculo + usuario
  segundo_anio = primer_anio + calculo
  tercer_anio = segundo_anio + calculo

  print(f'Interés por cada año es de 4%')
  print(f'Cantidad de ahorros del Primer Año  ${int(primer_anio)}')
  print(f'Cantidad de ahorros del Segundo Año ${int(segundo_anio)}')
  print(f'Cantidad de ahorros del Tercer Año  ${int(tercer_anio)}')

cuenta_ahorros()
```

Escribir una función que calcule el total de una factura tras aplicarle el IVA. 
La función debe recibir la cantidad sin IVA y el porcentaje de IVA a aplicar, y devolver el total de la factura. Si se 
invoca la función sin pasarle el porcentaje de IVA, deberá aplicar un 21%.

```python
# Funcion que recibe 2 parametros , iva trae valor por defecto si es que no ingresa ningun valor
def factura(cant, iva=21):
  # calculo del porcentaje 
  total_iva = (cant*iva/100) + cant


  print(f"""
  Total de Factura

  Cantidad  ${cant} 
  IVA       {iva}%
  T_pagar   ${total_iva}
  
  """)

#variable costo que recibe a la funcion de VALIDACION, esta validando solo numeros
costo = validacion('Ingrese valor a pagar: ')
iva = input('Desea ingresa IVA ?  SI / NO: ').lower()
# condicion que evalua si es que sera con IVA o no
if iva == 'si':
  iva_c = validacion('Ingrese valor de IVA: ')
  factura(costo, iva_c)
else:
  factura(costo)
```
